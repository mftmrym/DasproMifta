<h3>Data Mahasiswa</h3>

```python
import pandas as pd
import numpy as np
import random

# Jumlah baris data
jumlah_mahasiswa = 10000

# Data untuk disimulasikan
nama_depan = ['Adi', 'Budi', 'Citra', 'Dina', 'Eka', 'Fajar', 'Gita', 'Hari', 'Indah', 'Joko']
nama_belakang = ['Santoso', 'Wijaya', 'Putri', 'Saputra', 'Nugroho', 'Wibowo']
mata_kuliah = ['Struktur Data', 'Algoritma', 'Kecerdasan Buatan', 'Jaringan Komputer', 'Pemrograman Web', 'Sistem Operasi']
asal_daerah = ['Jakarta', 'Bandung', 'Surabaya', 'Yogyakarta', 'Medan', 'Makassar']

# Generate Data
np.random.seed(42)  # reproducibility
data = {
    'NIM': ['IF' + str(i).zfill(6) for i in range(1, jumlah_mahasiswa + 1)],
    'Nama': [random.choice(nama_depan) + ' ' + random.choice(nama_belakang) for _ in range(jumlah_mahasiswa)],
    'Jenis Kelamin': np.random.choice(['L', 'P'], jumlah_mahasiswa),
    'Umur': np.random.randint(17, 26, jumlah_mahasiswa),
    'Tahun Masuk': np.random.choice([2019, 2020, 2021, 2022, 2023], jumlah_mahasiswa),
    'IPK': np.round(np.random.normal(3.0, 0.4, jumlah_mahasiswa), 2),
    'Mata Kuliah Favorit': np.random.choice(mata_kuliah, jumlah_mahasiswa),
    'Keaktifan Organisasi': np.random.randint(0, 11, jumlah_mahasiswa),
    'Status Beasiswa': np.random.choice([True, False], jumlah_mahasiswa, p=[0.3, 0.7]),
    'Status Kelulusan': np.random.choice(['Lulus', 'Belum Lulus'], jumlah_mahasiswa, p=[0.4, 0.6]),
    'Jam Belajar per Hari': np.round(np.random.normal(2.5, 1.0, jumlah_mahasiswa), 2),
    'Domisili': np.random.choice(asal_daerah, jumlah_mahasiswa)
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Memastikan nilai IPK dan Jam Belajar berada pada rentang yang valid
df['IPK'] = df['IPK'].clip(lower=0.0, upper=4.0)
df['Jam Belajar per Hari'] = df['Jam Belajar per Hari'].clip(lower=0.0)

# Menampilkan informasi dasar
print("Head of Data:")
print(df.head())

print("\nInfo:")
print(df.info())

print("\nDescriptive Statistics:")
print(df.describe(include='all'))

# Contoh Karakteristik Data
print("\nDistribusi Jenis Kelamin:")
print(df['Jenis Kelamin'].value_counts())

print("\nDistribusi Mata Kuliah Favorit:")
print(df['Mata Kuliah Favorit'].value_counts())

print("\nRata-rata IPK per Tahun Masuk:")
print(df.groupby('Tahun Masuk')['IPK'].mean())
```
